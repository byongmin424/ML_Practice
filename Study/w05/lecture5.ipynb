```python
# test_4
# 경사하강법 프로그램을 이용해 최적 매개변수를 구하라.
# 단, 경사하강법 외부 함수 사용 금지!
# 단, 학습률, 초기값, 반복 회수는 임의로 정하여 사용!
# 결과물: 코드
# 결과물: 학습률, 초기값, 반복 횟수, 최종 평균제곱오차, 최적 매개변수

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("multiple_linear_regression_data.csv", names= ['height','weight','age'])
height = data['height'].to_numpy()
weight = data['weight'].to_numpy()
age = data['age'].to_numpy()

X = np.c_[height, weight, np.ones(len(height))]
y = age.reshape(len(age),1)

#예측을 위한 grid 생성 ################################################################
height_data = np.linspace(55, 190,1000)
weight_data = np.linspace(10, 100,1000)
Height, Weight = np.meshgrid(height_data, weight_data)

# analytic_y 해석해 ################################################################
analytic_W = np.linalg.pinv(X.T @ X) @ X.T @ y
analytic_y = analytic_W[0]*Height + analytic_W[1]*Weight + analytic_W[2]

# MSE by analytic_W ###############################################################
def function_MSE(W,X,y):
    predict = X.dot(W)
    error = predict - y
    mse = (error ** 2).mean()
    return mse

# grad ############################################################################
def numerical_gradient(f,W,X,y):
    h = 1e-4
    grad = np.zeros_like(W)
    
    for idx in range(W.size):
        tmp_val = W[idx]
        # f(x+h)
        W[idx] = tmp_val + h
        fxh1 = f(W, X, y)
        # f(x-h)
        W[idx] = tmp_val - h
        fxh2 = f(W, X, y)
        
        grad[idx] = (fxh1 - fxh2) / (2 * h)
        W[idx] = tmp_val
    
    return grad
####################################################################################

# 경사하강법 실행 ####################################################################
num_epoch = 100_000
eta = 1e-5
errors = []
GD_W0_history = []
GD_W1_history = []
GD_W2_history = []

GD_W = np.zeros((3,1))
GD_W[0] = np.random.uniform(low=-1.0, high=1.0)
GD_W[1] = np.random.uniform(low=-1.0, high=1.0)
GD_W[2] = np.random.uniform(low=-13.0, high=-10.0)
print(f'random w: \n{GD_W}')


for epoch in range(num_epoch):
    mse = function_MSE(GD_W, X, y)
    gradient = numerical_gradient(function_MSE, GD_W, X, y)
    GD_W -= eta * gradient
    
    if mse < 1.7:
        print('GD end')
        break
    
    errors.append(mse)
    GD_W0_history.append(GD_W[0])
    GD_W1_history.append(GD_W[1])
    GD_W2_history.append(GD_W[2])

    if epoch%100 == 0:
        print(f'epoch:{epoch} ===> W: {GD_W.flatten()}, gradient:{gradient.flatten()}, mse:{mse}')
    
GD_y = GD_W[0]*Height + GD_W[1]*Weight + GD_W[2]

####################################################################################

fig = plt.figure(1)
ax = fig.add_subplot(projection='3d')
ax.scatter(height, weight, age)
ax.plot_surface(Height, Weight, analytic_y, cmap='plasma')
ax.set_xlabel('height')
ax.set_ylabel('weight')
ax.set_zlabel('age')
ax.set_title('analytic_y')
plt.show()

fig = plt.figure(2)
ax = fig.add_subplot(projection='3d')
ax.scatter(height, weight, age)
ax.plot_surface(Height, Weight, GD_y, cmap='plasma')
ax.set_xlabel('height')
ax.set_ylabel('weight')
ax.set_zlabel('age')
ax.set_title('GD_y')
plt.show()
```

```python
# test_5
# 주어진 데이터에 대해 K개의 가우스 함수를 이용한 선형 기저함수 회귀모델의 최적 매개변수(해석해)를 자동 계산하는 프로그램을 
#   작성하고, K가 3,5,8 일 때의 매개변수를 구하라.
# 결과물 : 코드, 매개변수 
 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('linear_regression_data01.csv', names= ['age','height'])

age = data['age'].to_numpy()
X = age
X = np.c_[X, np.ones(len(X))]
height = data['height'].to_numpy()
y = height
y = y.reshape((len(y),1))

# 가우시안 기저함수 정의
def gaussian_basis_function(x, mu, sigma):
    return np.exp(-0.5 * ((x-mu)/sigma)**2)

# 선형 기저함수 모델의 해석해
def analytic_linear_W(X,y,mus,sigma):
    # 설계 행렬 생성
    Z = np.column_stack([gaussian_basis_function(X[:,0], mu, sigma) for mu in mus])
    # 1의 열을 추가하여 절편을 포함
    Z = np.c_[Z, np.ones(len(Z))]
    # 정규 방정식을 이용한 가중치 계산
    W = np.linalg.inv(Z.T @ Z) @ Z.T @ y
    return W

def calculate_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean

# 가우스 기저함수의 중심(mu)들과 sigma 설정
k_value = [3,5,8]
linear_W_result = {}

for k in k_value:
    mus = np.linspace(min(age),max(age),k)
    sigma = (max(age) - min(age)) / k
    linear_W = analytic_linear_W(X,y,mus,sigma)
    linear_W_result[k] = linear_W
    
    

    # 최적 매개변수 결과 출력
    print(f"k = {k}, Computed weights:\n{linear_W}")
```

```python
# test_6
# K = 3, 5, 8, 10일 떄, 훈련 데이터와 선형 기저함수 회귀 모델을 그래프에 표시하라.
# 결과물 : 그래프

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read data
raw_data = pd.read_csv("linear_regression_data01.csv", names = ['age','height'])

age = raw_data['age'].to_numpy()
height = raw_data['height'].to_numpy()
X = age
y = height

X = np.c_[X, np.ones(len(X))]
y = y.reshape(len(y),1)

# 주어진 age 값에 대해 회귀 모델의 예측값을 계산하는 함수
def predict(age, W, mus, s):
    Phi = np.column_stack([gaussian_basis_function(age, mu, s) for mu in mus] + [np.ones(len(age))])
    return Phi @ W

# 가우스 기저 함수 정의
def gaussian_basis_function(x, mu, s):
    return np.exp(-0.5 * ((x - mu) ** 2) / s ** 2)

# 선형 기저 함수 회귀 모델을 위한 최적 매개변수 계산 함수
def compute_optimal_parameters(K, age, y):
    # K개의 기저 함수 중심(mu)를 age의 최소값과 최대값 사이에 균등하게 위치
    mus = np.linspace(age.min(), age.max(), K)
    
    # 기저 함수 파라미터 설정
    s = (age.max() - age.min()) / K  # 표준편차를 데이터 범위에 따라 조정
    
    # 기저 함수 행렬(Phi) 구성
    Phi = np.column_stack([gaussian_basis_function(age, mu, s) for mu in mus] + [np.ones(len(age))])
    
    # 최적 매개변수(가중치) 계산
    W = np.linalg.pinv(Phi.T @ Phi) @ Phi.T @ y
    
    return W

# MSE를 계산하는 함수
def calculate_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# 훈련 데이터와 선형 기저함수 회귀 모델을 그래프에 표시하고 MSE를 계산하는 함수
def plot_model_with_mse(K, age, y):
    mus = np.linspace(age.min(), age.max(), K)
    s = (age.max() - age.min()) / K
    W = compute_optimal_parameters(K, age, y)
    
    # 예측값 계산
    predictions = predict(age, W, mus, s)
    
    # MSE 계산
    mse = calculate_mse(y, predictions)
    
    # 예측값을 더 부드럽게 하기 위한 그리드 생성
    age_grid = np.linspace(age.min(), age.max(), 100)
    predictions_grid = predict(age_grid, W, mus, s)
    
    plt.figure(figsize=(10, 6))
    plt.plot(age_grid, predictions_grid, color='red', label=f'predict, k={K}, MSE: {mse:.5f}')
    plt.scatter(age, y, label='original')
    plt.xlabel('age'); plt.ylabel('height')
    plt.title('Regression with gaussian basis function ')
    plt.legend(); plt.grid()
    plt.show()

# K = 3, 5, 8, 10에 대한 모델 그래프 및 MSE 출력
for K in [3, 5, 8, 10]:
    plot_model_with_mse(K, age, y)
```

```python
# test_7
# 실습 #6에서 각 K개에 대한 평균제곱오차를 구하고 x축은 K값, y축은
#    평균제곱오차를 나타내는 2차원 그래프를 구하라.
# 필수요소: x축, y축 이름, grid, legend
# 결과물: 코드, 그래

# read data
raw_data = pd.read_csv("linear_regression_data01.csv", names = ['age','height'])

age = raw_data['age'].to_numpy()
height = raw_data['height'].to_numpy()

X = np.c_[age, np.ones(len(age))]
y = height.reshape(len(height),1)

# 가우스 기저 함수 정의
def gaussian_basis_function(x, mu, s):
    return np.exp(-0.5 * ((x - mu) ** 2) / s ** 2)

# 선형 기저 함수 회귀 모델을 위한 최적 매개변수 계산 함수
def compute_optimal_parameters(K, age, y):
    # K개의 기저 함수 중심(mu)를 age의 최소값과 최대값 사이에 균등하게 위치
    mus = np.linspace(age.min(), age.max(), K)
    
    # 기저 함수 파라미터 설정
    s = (age.max() - age.min()) / K  # 표준편차를 데이터 범위에 따라 조정
    
    # 기저 함수 행렬(Phi) 구성
    Phi = np.column_stack([gaussian_basis_function(age, mu, s) for mu in mus] + [np.ones(len(age))])
    
    # 최적 매개변수(가중치) 계산
    W = np.linalg.pinv(Phi.T @ Phi) @ Phi.T @ y
    
    return W

# 주어진 age 값에 대해 회귀 모델의 예측값을 계산하는 함수
def predict(age, W, mus, s):
    Phi = np.column_stack([gaussian_basis_function(age, mu, s) for mu in mus] + [np.ones(len(age))])
    return Phi @ W

# MSE를 계산하는 함수
def calculate_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# K값 리스트 정의
K_values = [3, 5, 8, 10]

# MSE 값을 저장할 리스트 초기화
mse_values = []

# 각 K값에 대한 MSE 계산
for K in K_values:
    mus = np.linspace(age.min(), age.max(), K)
    s = (age.max() - age.min()) / K
    W = compute_optimal_parameters(K, age, y)
    
    # 예측값 계산
    predictions = predict(age, W, mus, s)
    
    # MSE 계산 및 저장
    mse = calculate_mse(y, predictions)
    mse_values.append(mse)

# MSE 그래프 그리기
plt.figure(figsize=(10, 6))
plt.stem(K_values, mse_values, label='MSE')
plt.xlabel('K')
plt.ylabel('MSE')
plt.grid(True); plt.legend()
plt.show()
```
